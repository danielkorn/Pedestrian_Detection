```python
class Vehicle(object):
    def __init__(self, xml_element):
        poses = xml_element[5]
        frames = poses[2:]
        
        self.object_type = xml_element[0].text
        self.h = xml_element[1].text
        self.w = xml_element[2].text
        self.l = xml_element[3].text
        self.first_frame = int(xml_element[4].text)
        self.num_frames = int(poses[0].text)
        
        self.frames = dict(enumerate([Frame(frame) for frame in frames]))
    
    def get_frame(self, i):
        return self.frames[i - self.first_frame]
    
class Frame(object):
    def __init__(self, xml_element):
        self.tx = xml_element[0].text
        self.ty = xml_element[1].text
        self.tz = xml_element[2].text
        self.rx = xml_element[3].text
        self.ry = xml_element[4].text
        self.rz = xml_element[5].text
        self.state = xml_element[6].text
        self.occlusion = xml_element[7].text
        self.occlusion_kf = xml_element[8].text
        self.truncation = xml_element[9].text
        self.amt_occlusion = xml_element[10].text
        self.amt_occlusion_kf = xml_element[11].text
        self.amt_border_l = xml_element[12].text
        self.amt_border_r = xml_element[13].text
        self.amt_border_kf = xml_element[14].text
```
